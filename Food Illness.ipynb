```python
import pandas as pd
import numpy as np

df = pd.read_csv('outbreaks.csv')
df.head()
```

```python
food = df[df['Food'].notnull()]
ingredient = df[df['Ingredient'].notnull()]
species = df[df['Species'].notnull()]
sg = df[df['Serotype/Genotype'].notnull()]
```

```python
#Are foodborne disease outbreaks increasing or decreasing? 
pd.pivot_table(df, index = ['Year'], values=['Illnesses'], aggfunc='sum')
#looks like has decreased somewhat
```

```python
#What contaminant has been responsible for the most illnesses, hospitalizations, and deaths? 
foodpt = pd.pivot_table(food, index = ['Food'], values=['Illnesses','Hospitalizations','Fatalities'], aggfunc='sum')
foodpt.sort_values(['Illnesses'], ascending=[False], inplace=True)
foodpt
#multiple foods, sald, pork, tomato
```

```python
foodpt.sort_values(['Hospitalizations'], ascending=[False], inplace=True)
foodpt
#cantelope
```

```python
foodpt.sort_values(['Fatalities'], ascending=[False], inplace=True)
foodpt
#cantelope
```

```python
#cantalope is a bad food, can be hospitalized and die if eat bad cantelope
```

```python
ingredientpt = pd.pivot_table(ingredient, index = ['Ingredient'], values=['Illnesses','Hospitalizations','Fatalities'], aggfunc='sum')
ingredientpt.sort_values(['Illnesses'], ascending=[False], inplace=True)
ingredientpt
#egg not good
```

```python
ingredientpt.sort_values(['Hospitalizations'], ascending=[False], inplace=True)
ingredientpt
#egg
```

```python
ingredientpt.sort_values(['Fatalities'], ascending=[False], inplace=True)
ingredientpt
#cantelope
```

```python
speciespt = pd.pivot_table(species, index = ['Species'], values=['Illnesses','Hospitalizations','Fatalities'], aggfunc='sum')
speciespt.sort_values(['Illnesses'], ascending=[False], inplace=True)
speciespt
#Norovirus genogroup I
```

```python
speciespt.sort_values(['Hospitalizations'], ascending=[False], inplace=True)
speciespt
#Salmonella enterica
```

```python
sgpt = pd.pivot_table(sg, index = ['Serotype/Genotype'], values=['Illnesses','Hospitalizations','Fatalities'], aggfunc='sum')
sgpt.sort_values(['Illnesses'], ascending=[False], inplace=True)
sgpt
#Unknown
```

```python
#What location for food preparation poses the greatest risk of foodborne illness?
df["Location"].fillna('Unknown', inplace=True)

df.loc[df['Location'].str.contains("Restaurant"),'LocationCategory'] ='Restaurant'
df.loc[df['Location'].str.contains("Home"),'LocationCategory'] ='Home'
df.loc[df['Location'].str.contains("School"),'LocationCategory'] ='School'
df.loc[df['Location'].str.contains("Office"),'LocationCategory'] ='Office'
df.loc[df['Location'].str.contains("Banquet"),'LocationCategory'] ='Banquet'
df.loc[df['Location'].str.contains("Catering Service"),'LocationCategory'] ='Catering Service'
df.loc[df['Location'].str.contains("Fair"),'LocationCategory'] ='Fair'
df.loc[df['Location'].str.contains("Grocery"),'LocationCategory'] ='Grocery'
df.loc[df['Location'].str.contains("Boat"),'LocationCategory'] ='Boat'
df.loc[df['Location'].str.contains('Hotel'),'LocationCategory'] ='Hotel'
df.loc[df['Location'].str.contains("Jail"),'LocationCategory'] ='Jail'
df.loc[df['Location'].str.contains("Nursing"),'LocationCategory'] ='Nursing'
df.loc[df['Location'].str.contains("Camp"),'LocationCategory'] ='Camp'
df.loc[df['Location'].str.contains("Religious"),'LocationCategory'] ='Religious'
df.loc[df['Location'].str.contains("Other"),'LocationCategory'] ='Other'
df.loc[df['Location'].str.contains("Uknown"),'LocationCategory'] ='Uknown'
df.loc[df['Location'].str.contains("Daycare"),'LocationCategory'] ='Daycare'
df.head()
```

```python
#location 
location = pd.pivot_table(df, index = ['LocationCategory','Location'], values=['Illnesses','Hospitalizations','Fatalities'], aggfunc='sum')
location.sort_values(['Illnesses'], ascending=[False], inplace=True)
location
#Restaurant
```

```python
location.sort_values(['Hospitalizations'], ascending=[False], inplace=True)
location
#Restaurant
```

```python
location.sort_values(['Fatalities'], ascending=[False], inplace=True)
location
#Home
```
